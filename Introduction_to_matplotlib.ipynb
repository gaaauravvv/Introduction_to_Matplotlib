```python
#importing matplotlib librabry
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#defining the coordinates to be plot
x=[1,2,3,4,5,6,7,8,9]
y=[23,43,12,90,6,54,56,77,80]


```

```python
#giving labels to x axis,y axis and title of the graph
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
```

```python
#plotting the values
plt.plot(x,y)
```

```python
#plotting with different colors and linestylr
plt.plot(x,y,color='red',linewidth=2,linestyle='dashed')
```

```python
#plot with circle at values and rest connecting with '--'
plt.plot(x,y,'go--')
```

```python
#defining marker
plt.plot(x,y,color='red',marker='+',linestyle='dashed')
```

# defining size of marker


```python
plt.plot(x,y,color='blue',marker='+',markersize=10,linestyle='dashed')
```

# alpha property to define transparency

```python
plt.plot(x,y,'y^',alpha=0.9,linestyle='dashed') #alpha is b/w 0 and 1 both values included
```

# axes labels, legend and grid

```python
#matplotlib is already imported hence we directly define the values to be plotted
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label='max',color='blue')
plt.plot(days,min_t,label='min', color='red')
plt.plot(days,avg_t,label='avg',color='yellow')
```

# axes label and chart title

```python
#set axes lables and title of the chart
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

#plotting the graph again
plt.plot(days,max_t,label='max',color='blue')
plt.plot(days,min_t,label='min', color='red')
plt.plot(days,avg_t,label='avg',color='yellow')
```

# legend

```python
#showing legend
plt.plot(days,max_t,label='max',color='blue')
plt.plot(days,min_t,label='min', color='red')
plt.plot(days,avg_t,label='avg',color='yellow')

plt.legend(loc='best',shadow=True,fontsize=12) #places the legend on the best suitible position,shows shadow and set the font size of the legend
```
